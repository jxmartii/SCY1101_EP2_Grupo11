### Objetivo:
Optimizar el modelo seleccionado (Random Forest) mediante una búsqueda en cuadrícula combinada con Validación Cruzada (`GridSearchCV`). El fin es sintonizar los hiperparámetros del algoritmo para mejorar su capacidad de generalización en la predicción del `Rango_precio` en la región de Arica y Parinacota.

```python
import os
import joblib
import pandas as pd
import numpy as np

# 1. Configuración de rutas para Google Colab si corresponde
if 'google.colab' in str(get_ipython()):
    if not os.path.exists('SCY1101_EP2_Grupo11'):
        !git clone https://github.com/jxmartii/SCY1101_EP2_Grupo11.git
    os.chdir('/content/SCY1101_EP2_Grupo11/notebooks')

# 2. Cargar los datos de entrenamiento y el modelo baseline
X_train, X_test, y_train, y_test = joblib.load('../models/datos_entrenamiento.pkl')
modelo_base_rf = joblib.load('../models/baseline_random_forest.pkl')

print("¡Datos cargados correctamente! Iniciando el proceso de optimización...")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# 1. Definir la malla de hiperparámetros a evaluar
param_grid = {
    'n_estimators': [50, 100],        # Número de árboles en el bosque
    'max_depth': [10, 20, None],       # Profundidad máxima de los árboles
    'min_samples_split': [2, 5]        # Mínimo de muestras para dividir un nodo
}

# 2. Instanciar el optimizador usando Validación Cruzada de 3 pliegues (3-Fold CV)
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=3,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# 3. Ejecutar la búsqueda con los datos de entrenamiento
print("Entrenando combinaciones con GridSearchCV...")
grid_search.fit(X_train, y_train)

# 4. Desplegar los mejores parámetros encontrados
print("\n=== RESULTADOS DE OPTIMIZACIÓN ===")
print(f"Mejores hiperparámetros encontrados: {grid_search.best_params_}")
print(f"Mejor Accuracy obtenido en validación cruzada: {grid_search.best_score_:.4f}")
```

```python
from sklearn.metrics import classification_report, accuracy_score

# 1. Extraer el mejor modelo ya entrenado automáticamente
modelo_optimizado_rf = grid_search.best_estimator_

# 2. Realizar predicciones sobre el conjunto de test independiente
preds_optimizadas = modelo_optimizado_rf.predict(X_test)

# 3. Mostrar el reporte final de clasificación
print("=== REPORTE DE CLASIFICACIÓN: RANDOM FOREST OPTIMIZADO ===")
print(f"Accuracy Final en Test: {accuracy_score(y_test, preds_optimizadas):.4f}\n")
print(classification_report(y_test, preds_optimizadas))
```

```python
# Guardar el modelo final optimizado en la carpeta /models
# Este archivo será el corazón de la última etapa (Notebook 05)
joblib.dump(modelo_optimizado_rf, '../models/modelo_final_random_forest.pkl')

print("¡Modelo final sintonizado e hiperparámetros guardados exitosamente en /models!")
print("Listo para exportar al análisis final de la entrega.")
```
